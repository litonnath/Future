```
#this is the link i have scraped above from this raajkart website
import pandas as pd
plink=pd.read_excel('/content/drive/MyDrive/plink.xlsx')
plink
```

```
#data will scraped from this
#i m going to scraped only 50 data ,it has 36 thousand products.
proxies={'http':'http://user:pass@172.28.0.2/'}
Data=[]
for j in range(1023,1050):
  # print(li['URL'][j])
  
  uclient=requests.get(plink[0][j],proxies=proxies)
  # page_html=uclient.read()
  # uclient.close()
  
  page_soup=soup(uclient.content,"html.parser")
  print (j)
  price=page_soup.find('p',{'class':'old-price'})
  
  title=page_soup.find('div',{'class':'product-name'})
 
  short=page_soup.find('div',{'class':'short-description'})
 
  des=page_soup.find('div',{'class':'add-to-box'})
 
  long=page_soup.find('div',{'class':'box-collateral-content'})
 
  img=[]
  for g in page_soup.find_all('p',{'class':'product-image'}):
    for a in g.find_all('a'):
      img.append(a['href'])
  ty=[]
  for tb in page_soup.find_all('table',{'class':'data-table'}):
    for tr in tb.find_all('tr'):
      d=[]
      for th,td in zip(tr.find_all('th'),tr.find_all('td')):
        d.append(th.text+":"+td.text)
      ty.append(d)
  Data.append((plink[0][j],price,title,short,des,long,img,ty))
```

```
import pandas  as pd
h=pd.DataFrame(Data,columns=['URL','PRICE','INFO','SHORTDESC','DETAILS','LONGDESC','IMAGEURL','PRODUCTTABLE'])
h
```

```
h.to_excel('1000to2028raajkart.xlsx')
```
