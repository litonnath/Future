```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup as bs
import csv
import requests
```

```python
url = "http://www.westinghouselighting.com/light-bulbs/led-bulbs/reflector/6-watt-(replaces-40-watt)-par16-reflector-dimmable-led-light-bulb-0306700.aspx"
response = requests.get(url)
print(response.status_code)
```

```python
soup = bs(response.text,"html.parser")
soup
```

```python
table = soup.findAll('table',{"class":"Spectbl"})[0]
values =[]

tr = table.findAll(['tr'])[0:18]
```

```python
csvFile = open("product.csv",'wt',newline='',encoding='utf-8')
writer = csv.writer(csvFile)  
try:   
        for cell in tr:
            th = cell.find_all('th')
            th_data = [col.text.strip('\n') for col in th]
            td = cell.find_all('td')
            row = [i.text.replace('\n','') for i in td]
            writer.writerow(th_data+row)      
        
finally:   
    csvFile.close()
```
